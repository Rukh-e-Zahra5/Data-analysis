## Dataset Description

* **Political Party:** Names of political parties.
* **Province:** Names of provinces (Punjab, Sindh, KPK, Balochistan).
* **Seats Won:** Number of seats won by each party in each province.

```python
import pandas as pd

# Simulated dataset for Pakistan elections results
data = {
    'Political_Party': ['PTI', 'PML-N', 'PPP', 'ANP', 'MQM', 'Independent'],
    'Punjab': [120, 160, 20, 0, 0, 40],
    'Sindh': [30, 10, 90, 0, 20, 5],
    'KPK': [75, 20, 10, 5, 0, 5],
    'Balochistan': [10, 15, 5, 0, 0, 10]
}

df = pd.DataFrame(data)
```

## Exploratory Data Analysis (EDA)

We will inspect the data before moving to visualizations.

```python
# Display the dataset
print(df)

# Basic summary statistics
print(df.describe())
```

## Visualizations Using Matplotlib and Seaborn

A. **Bar Plot** - Seats Won by Each Political Party in Each Province

```python
import matplotlib.pyplot as plt

# Set Political Party as index for easier plotting
df.set_index('Political_Party', inplace=True)

# Bar plot showing the seats won by each party in all provinces
df.plot(kind='bar', stacked=False, figsize=(10,6))
plt.title('Seats Won by Political Parties in Each Province')
plt.xlabel('Political Party')
plt.ylabel('Number of Seats')
plt.legend(title='Province')
plt.show()
```

B. **Pie Chart** - Total Seats Distribution by Party

```python
# Sum of seats won by each party across all provinces
total_seats = df.sum(axis=1)

# Plot pie chart
plt.figure(figsize=(8,8))
plt.pie(total_seats, labels=total_seats.index, autopct='%1.1f%%', startangle=90)
plt.title('Total Seats Distribution by Political Party')
plt.show()
```

C. **Dot Plot** - Seats Won in Each Province by Party

```python
import seaborn as sns

# Prepare data for the dot plot
df_reset = df.reset_index().melt(id_vars='Political_Party', var_name='Province', value_name='Seats_Won')

# Dot plot
plt.figure(figsize=(10,6))
sns.stripplot(x='Seats_Won', y='Province', hue='Political_Party', data=df_reset, size=10, jitter=True)
plt.title('Seats Won in Each Province by Political Party')
plt.xlabel('Number of Seats')
plt.ylabel('Province')
plt.legend(title='Political Party', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

## Documenting Findings and Insights

**Insights from the Bar Plot:**

* PML-N dominates in Punjab, winning the highest number of seats.
* PPP leads in Sindh with the majority of seats, while PTI leads in KPK.

**Insights from the Pie Chart:**
* PML-N holds the largest share of total seats across all provinces.
* PPP and PTI follow, with Independent candidates taking a smaller portion.

**Insights from the Dot Plot:**
* The dot plot gives a more detailed view of seat distribution in each province.
* It highlights the variability in performance across different regions, especially showing parties like MQM focusing primarily on urban Sindh, while ANP has a small presence in KPK.
